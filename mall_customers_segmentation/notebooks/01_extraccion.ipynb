```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("vjchoudhary7/customer-segmentation-tutorial-in-python")

print("Path to dataset files:", path)
```

```python
import os
import zipfile

# 'path' variable exists from previous cell
if os.path.exists(path):
  print("Dataset already available at:", path)
else:
  # download (kagglehub is already imported in the notebook)
  path = kagglehub.dataset_download("vjchoudhary7/customer-segmentation-tutorial-in-python")
  print("Downloaded dataset to:", path)

# show contents
print("\nFiles in dataset directory:")
for root, dirs, files in os.walk(path):
  for fname in files:
    print(os.path.join(root, fname))

# extract any zip files found in the dataset directory
for fname in os.listdir(path):
  if fname.lower().endswith(".zip"):
    zip_path = os.path.join(path, fname)
    extract_dir = os.path.join(path, os.path.splitext(fname)[0])
    os.makedirs(extract_dir, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as z:
      z.extractall(extract_dir)
    print(f"Extracted {zip_path} -> {extract_dir}")
```

```python
# load files to data
import pandas as pd
data_file = os.path.join(path, "Mall_Customers.csv")
data = pd.read_csv(data_file)
print("\nData preview:")
print(data.head())
```

```python
data.to_csv("../data/mall_customers.csv", index=False)
```
